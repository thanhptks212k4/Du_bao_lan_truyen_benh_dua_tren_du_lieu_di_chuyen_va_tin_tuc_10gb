```python
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ==== 0. FIX SEED ====
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

# ==== 1. ĐỌC & TIỀN XỬ LÝ DỮ LIỆU ====
df = df.drop_duplicates(subset=['username', 'text']).reset_index(drop=True)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
daily_counts = df.groupby(df['date'].dt.date).size().reset_index(name='post_count')
daily_counts['date'] = pd.to_datetime(daily_counts['date'])
daily_counts = daily_counts.sort_values('date').reset_index(drop=True)

# Log transform + MinMaxScaler
scaler = MinMaxScaler()
daily_counts['log_post_count'] = np.log1p(daily_counts['post_count'])
daily_counts['scaled_post_count'] = scaler.fit_transform(daily_counts[['log_post_count']])

# Gán nhãn phân loại theo phân vị của post_count gốc
q25 = daily_counts['post_count'].quantile(0.25)
q75 = daily_counts['post_count'].quantile(0.75)

def label(x):
    if x <= q25:
        return 0
    elif x >= q75:
        return 2
    else:
        return 1

daily_counts['label'] = daily_counts['post_count'].apply(label)

# ==== 2. TẠO INPUT/OUTPUT CHO DỰ BÁO ====
sequence_length = 30
X, y_cls, y_val = [], [], []
for i in range(len(daily_counts) - sequence_length):
    seq = daily_counts['scaled_post_count'].iloc[i:i + sequence_length].values
    target_val = daily_counts['scaled_post_count'].iloc[i + sequence_length]
    target_cls = daily_counts['label'].iloc[i + sequence_length]
    X.append(seq)
    y_val.append(target_val)
    y_cls.append(target_cls)

X = np.array(X).astype(np.float32)
y_val = np.array(y_val).astype(np.float32).reshape(-1, 1)
y_cls = np.array(y_cls).astype(np.int64)

# ==== 3. MÔ HÌNH PATCHTST ====
class PatchTST(nn.Module):
    def __init__(self, patch_len, input_len, d_model=64, n_heads=4, num_classes=3):
        super(PatchTST, self).__init__()
        self.patch_len = patch_len
        self.num_patches = input_len // patch_len
        self.embed = nn.Linear(patch_len, d_model)

        self.encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model, nhead=n_heads, dim_feedforward=128, batch_first=True
        )
        self.encoder = nn.TransformerEncoder(self.encoder_layer, num_layers=2)
        self.cls_head = nn.Linear(d_model, num_classes)

    def forward(self, x):
        B, L = x.shape
        x = x.view(B, self.num_patches, self.patch_len)
        x = self.embed(x)
        x = self.encoder(x)
        x = x.mean(dim=1)
        return self.cls_head(x)

# ==== 4. TRAIN/TEST SPLIT ====
X_train, X_test, y_cls_train, y_cls_test, y_val_train, y_val_test = train_test_split(
    X, y_cls, y_val, shuffle=False, test_size=0.2
)

# ==== 5. HUẤN LUYỆN MÔ HÌNH ====
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = PatchTST(patch_len=5, input_len=30, d_model=64, n_heads=4).to(device)

X_train_tensor = torch.tensor(X_train).to(device)
y_train_tensor = torch.tensor(y_cls_train).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

losses = []
for epoch in range(150):
    model.train()
    optimizer.zero_grad()
    output = model(X_train_tensor)
    loss = criterion(output, y_train_tensor)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1}, Loss: {loss.item():.6f}")

# ==== 6. ĐÁNH GIÁ ====
model.eval()
X_test_tensor = torch.tensor(X_test).to(device)
with torch.no_grad():
    logits = model(X_test_tensor)
    preds = torch.argmax(logits, dim=1).cpu().numpy()

y_cls_test = np.array(y_cls_test)
acc = accuracy_score(y_cls_test, preds)
print(f"\n  Accuracy: {acc*100:.2f}%\n")
print(classification_report(y_cls_test, preds, labels=[0,1,2], target_names=['Thấp', 'Trung bình', 'Cao']))
print(confusion_matrix(y_cls_test, preds))

# ==== 7. BIỂU ĐỒ LOSS ====
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Cross-Entropy Loss")
plt.title("Training Loss")
plt.grid(True)
plt.show()
```
